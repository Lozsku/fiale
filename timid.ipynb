```python
import shutil
import os
```

```python
for a in train_path:
    # Path to the file you want to copy
    source_file_path = 'data/'+a

    # Path to the directory where you want to paste the file
    destination_directory = 'train_data/'

    # Create the destination directory if it doesn't exist
    os.makedirs(destination_directory, exist_ok=True)

    # Construct the destination file path
    destination_file_path = os.path.join(destination_directory, os.path.basename(source_file_path))

    # Copy the file
    shutil.copy2(source_file_path, destination_file_path)
```

```python
for a in test_path:
    # Path to the file you want to copy
    source_file_path = 'data/'+a

    # Path to the directory where you want to paste the file
    destination_directory = 'test_data/'

    # Create the destination directory if it doesn't exist
    os.makedirs(destination_directory, exist_ok=True)

    # Construct the destination file path
    destination_file_path = os.path.join(destination_directory, os.path.basename(source_file_path))

    # Copy the file
    shutil.copy2(source_file_path, destination_file_path)
```
